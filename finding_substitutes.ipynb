```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
pd.set_option('display.max_rows', None)
```

```python
clusters_df = pd.read_csv('clusters_df.csv', header=None)
df =  pd.read_csv('mma_mart.csv')
```

```python
# Group by 'product_name' and sum rows for sales count
df_quantity_sold = df.groupby('product_name').size().reset_index(name='total_quantity_sold')

# Merge with df to carryover product details
df_quantity_sold = pd.merge(df_quantity_sold, df.drop_duplicates('product_name'), on='product_name')
df_quantity_sold = df_quantity_sold.sort_values(by='total_quantity_sold', ascending=False)
df = df_quantity_sold
```

```python
cluster_max_product = {}
for column in clusters_df.iloc[:, :0].columns:
    products_in_cluster = clusters_df[column]
```

```python
products_in_cluster.head(100)
```

```python
cluster_max_product = {}
for column in clusters_df.columns:
    products_in_cluster = clusters_df[column] # Products in each cluster
    if not products_in_cluster.empty:
        mask = df['product_name'].isin(products_in_cluster) # Boolean mask if each row is in cluster
        if any(mask):  # Check if there are any products in the cluster
            max_product = df.loc[mask, 'total_quantity_sold'].idxmax()
            top_seller = df.loc[max_product, 'product_id']
            cluster_max_product[column] = top_seller
            df.loc[mask, 'substitute'] = top_seller
```

```python
df['frozen'] = df['department'] == 'frozen'
dept_refrigerated = ['dairy eggs', 'deli', 'meat seafood', 'beverages_refrigerated']
df['refrigerated'] = df['department'].isin(dept_refrigerated)
```
